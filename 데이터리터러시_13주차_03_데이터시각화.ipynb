## seaborn으로 데이터 시각화하기

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(data = iris)    # 수치데이터를 두 데이터씩 짝 지어서 산점도 / 같은 데이터인 경우 히스토그램
```

```python
sns.pairplot(data = iris, hue = 'species')    # 종별로 색깔 구분
```

```python
sns.pairplot(data = iris, hue = 'species', vars = ['petal_length', 'petal_width'])    # 원하는 데이터 지정
```

```python
sns.pairplot(data = iris, hue = 'species', vars = ['petal_length', 'petal_width'], height = 4)    # 크기 지정
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'scatter')
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'reg')    # 회귀선
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'kde')    # 확률밀도
```

```python
sns.jointplot(x = 'petal_length', y = 'petal_width', data = iris, kind = 'hex')    # 육각형
```

```python
sns.lmplot(x = 'sepal_length', y = 'sepal_width', hue = 'species', data = iris)
```

```python
sns.lmplot(x = 'petal_length', y = 'petal_width', hue = 'species', data = iris)
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.heatmap(iris.corr(), annot = True)    # 수치 정보 제공
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Blues')
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Reds')
```

```python
sns.heatmap(iris.corr(), annot = True, cmap = 'Greens')
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', data = iris)
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', size = 'species', data = iris)    # 종별로 크기 조절
```

```python
sns.relplot(x = 'petal_length', y = 'petal_width', hue = 'species', data = iris)    # 산점도에서 종별 색깔 달리 표현
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'size')
```

```python
sns.pairplot(tips, hue = 'size', height = 4)
```

```python
sns.pairplot(tips, hue = 'day')
```

```python
sns.pairplot(tips, hue = 'day', height = 3)
```

```python
sns.barplot(x = 'day', y = 'tip', data = tips)
```

```python
tips['day'].value_counts()
```

```python
sns.barplot(x = 'day', y = 'tip', hue = 'smoker', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips)    # 사이즈별로 끊어서 그래프 그리기
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', col_wrap = 3, data = tips)    # 세 개씩 끊어서 보여달라
```

```python
pivot = tips.pivot_table(index = 'day', columns = 'smoker', values = 'tip')
```

```python
pivot
```

```python
sns.heatmap(pivot)
```

```python
sns.heatmap(pivot, annot = True)
```

```python
sns.heatmap(pivot, annot = True, cmap = 'Blues')
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', size = 'size', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', size = 'smoker', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', data = tips)
```
