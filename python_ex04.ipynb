```python
# 228 ~ 293 page 까지
def factorial(n):
    output = 1     # 곱셈 초기값 1
    for i in range(1, n+1):     # 1부터 n까지
        output *= i             # 모두 곱한다
    return output

print("1!:",factorial(1))
print("2!:",factorial(2))
print("3!:",factorial(3))
print("4!:",factorial(4))
print("5!:",factorial(5))
```

```python
# 재귀 함수로 팩토리얼 구하기

def factorial(n):
    if n == 0:     # n이 0이면 1 리턴
        return 1
    else:          # n이 0이 아니면 n * factorial(n-1)
        return n * factorial(n-1)
    
print("1!:",factorial(1))
print("2!:",factorial(2))
print("3!:",factorial(3))
print("4!:",factorial(4))
print("5!:",factorial(5))
```

```python
# 243페이지 1번 문제

def flatten(data):
    list_a = []                      # 모든 결과는 빈 리스트에 더해질 것이다.
    
    for i in data:
        if type(i) == list:
            list_a += flatten(i)     # 나온 결과가 각각 모두 더해져야 한다.
            
        else:
            list_a.append(i)         # 숫자형일 경우에는 빈 리스트에 더해진다.
            
    return list_a                   # 모든 계산이 끝난 후 list_a 반환

example = [[1, 2, 3], [4, [5, 6]], 7, [8, 9]]
print("원본:", example)
print("변환:", flatten(example))
```

```python
위의 문제는 사실상 모든 수를 리스트로 만들어서 리스트끼리 더하게 하는 모양이다.

example의 원소인 [1,2,3]과 [4,[5,6]]과 7 그리고 [8,9]를 각각 하나씩 리스트 형태인지 알아보기 위해서
flatten함수에 for 반복문을 설정하고, 하위에 if문을 설정하여 타입이 리스트일 경우와 리스트가 아닐 경우로 나눈다.

여기에서 중요한 것은 for 반복문 이전에 list_a = [], 즉 빈 리스트 공간을 만들어 놓는 것이다.
list_a는 이 문제에서 중요한 역할을 한다.

다시 if문으로 돌아가서, 타입이 리스트가 아닐 경우, 즉 평범한 숫자일 경우에는 빈 공간인 list_a에 그냥 더하기 위해 append를 사용한다.

문제는 if type(i) == list일 경우이다.
이 경우에는 리스트는 리스트끼리 더하여 표현할 수 있다는 특성을 생각해내야 한다.
예를 들어 a = [1,2,3]이고 b = [4,5,6]일 때 리스트인 a와 b를 더하면 a + b -> [1,2,3,4,5,6]이다.

따라서 빈 공간인 list_a에 리스트 형태를 더하는 형식을 취한다. list_a += 리스트
다만 여기에서 리스트는 또 안에 리스트를 원소로 가지고 있을 수 있으므로 "재귀함수"를 이용해야 한다.

example = [
    [1, 2, 3],      1번째
    [4, [5, 6]],    2번째
    7,              3번째
    [8, 9]          4번째
]

example이 flatten으로 들어갔다.
for문으로 1번째인 [1,2,3]이 들어갔다.
[1,2,3]은 리스트 형태이다. 따라서 list_a, 즉 []에 그냥 append할 수 없다. -> 다시 함수를 돌려야 한다.
    1,2,3이 각각 for문으로 들어갔다. 리스트형태가 아니므로 list_a에 각각 append된다. -> ## list_a -> [1,2,3]가 된다.

for문으로 2번째인 [4,[5,6]]이 들어갔다.
[4,[5,6]]은 리스트 형태이다. -> 다시 함수를 돌린다.
    4,[5,6]이 각각 for문으로 들어갔다. 4는 리스트형태가 아니어서 list_a에 append된다. -> ## list_a -> [4]가 된다.
                                       [5,6]은 리스트형태이다 -> 다시 함수를 돌린다.
                                            5,6이 각각 for문으로 들어간다. 5와 6은 리스트 형태가 아니다. -> ## list_a -> [5,6]
for문으로 3번째인 7이 들어갔다.
7은 리스트 형태가 아니다. list_a에 append된다. -> ## list_a -> [7]이 된다.

for문으로 4번째인 [8,9]가 들어갔다.
[8,9]는 리스트 형태이다 -> 다시 함수를 돌린다.
    8,9가 각각 for문으로 들어갔다. 둘은 리스트형태가 아니어서 list_a에 append된다. -> ## list_a -> [8,9]
    
각각 함수를 발동하며 얻은 list_a의 결과는 하나로 합쳐져야 한다.
그에 따라서 if문 하위에 list_a += flatten(i) 문장이 써진다.

결국 list_a += flatten(i)는
[] += [1,2,3] += ( [4] + [5,6] ) += ( [7] + [8,9] )된 것이다.
```

```python
# 244페이지 2번 문제

min = 2     # 테이블에 혼자 안돼! 적어도 두 명!
max = 10    # 테이블은 10명까지
every = 100 # 모든 사람 수
memo = {}   # 딕셔너리 - 여러 번 다시 계산하지 않기 위해 저장할 공간

def table(left, sat):     # 테이블에 어떻게 앉힐까? - 남은사람, 앉은사람
    key = str([left, sat])
    
    # 종료 조건
    if key in memo:       # 키값이 이미 저장되었다면
        return memo[key]  # 저장된 것을 반환
    if left < 0:
        return 0
    if left == 0:
        return 1
    
    # 재귀 처리
    count = 0                         # 초기값
    for i in range(sat, max + 1):    # 2명(sat=min)부터 10명(max)까지 범위를 잡음
        count += table(left - i, i)  # count += split(100명 - 2명, 2명)
    
    # 메모화 처리
    memo[key] = count
    
    # 종료
    return count

print(table(every, min))
```

```python
# filter() 함수와 map() 함수

# map(): 리턴된 값으로 새로운 리스트 구성
# filter(): 리턴된 값이 True인 것으로, 새로운 리스트 구성

def power(item):
    return item*item
def under_3(item):
    return item < 3

list_input_a = [1,2,3,4,5]

output_a = map(power, list_input_a)    # map(함수, 리스트)
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(under_3, list_input_a)
print("filter(under_3, list_input_a):", output_b)
print("filter(under_3, list_input_a):", list(output_b))
```

```python
# 제너레이터 - 이터레이터 직접 만들 때 사용하는 코드
# 함수 내부에 yield키워드 사용 -> 해당 함수는 제너레이터 함수가 됨
# 함수를 호출해도 함수 내부의 코드가 실행되지 않음

def test():
    print("함수가 호출되었습니다.")
    yield "test"
    
print("A 지점 통과")
test()                # 호출이 되지 않음

print("B 지점 통과")
test()                # 호출이 되지 않음
print(test())         # 프린트로 부르면 제너레이터 리턴
```

```python
# 제너레이터 객체는 next() 함수를 사용해 실행한다.
# yield 키워드 부분까지만 실행한다.
# next() 함수의 리턴값으로 yield 뒤에 입력한 값 출력

def test():
    print("A 지점 통과")
    yield 1
    print("B 지점 통과")
    yield 2
    print("C 지점 통과")
    
output = test()

print("D 지점 통과")
a = next(output)
print(a)
b = next(output)
print(b)
c = next(output)
print(c)          # yield 다음값까지 나와야 하는데 없으므로 예외 발생.

# next() 함수를 호출한 이후 
# yield 키워드를 만나지 못하고 함수가 끝나면
# StopIteration이라는 예외 발생
```

```python
def test():
    print("A 지점 통과")
    yield 1
    print("B 지점 통과")
    yield 2
    print("C 지점 통과")
    
output = test()       # 변수에 함수 호출값 할당
print("D 지점 통과")
a = next(output)      # 내부 코드 실행: 첫 번째 yield 값까지
print(a)
print("E 지점 통과")
b = next(output)      # 내부 코드 실행: 이어서 두 번째 yield 값까지
print(b)
print("F 지점 통과")
c = next(output)
print(c)

next(output)
```

### try except 구문

```python
# try: 예외가 발생할 가능성 있는 코드
# except: 예외가 발생했을 때 실행할 코드

try:
    number_input_a = int(input("정수 입력> ")) # 숫자로 변환
    
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2*3.14*number_input_a)
    print("원의 넓이:", 3.14*number_input_a*number_input_a)
except:
    print("무언가 잘못되었습니다.")
```

```python
# try except 구문과 pass 조합하기

#숫자로 변환되는 것들만 리스트에 넣기
list_input_a = ["52", "273", "32", "스파이", "103"]

list_number = []
for item in list_input_a:
    try:
        float(item) # 예외가 발생하면 알아서 다음으로 진행 안될 것
        list_number.append(item) # 통과했구나! 여기 들어와
    except:
        pass # 예외니까 다음으로 그냥 넘어가
    
print("{} 내부에 있는 숫자는".format(list_input_a))
print("{}입니다.".format(list_number))
```

```python
# try except else 구문

# else: 예외가 발생하지 않았을 때 실행할 코드

try:
    number_input_a = int(input("정수 입력> "))
except:
    print("정수를 입력하지 않았습니다.")
else:
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2*3.14*number_input_a)
    print("원의 넓이:", 3.14*number_input_a*number_input_a)
```

```python
# finally 구문
# finally: 무조건 실행할 코드

try:
    number_input_a = int(input("정수 입력> "))
    
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2*3.14*number_input_a)
    print("원의 넓이:", 3.14*number_input_a*number_input_a)
except:
    print("정수를 입력해달라고 했잖아요?!")
else:
    print("예외가 발생하지 않았습니다.")
finally:
    print("일단 프로그램이 어떻게든 끝났습니다.")
```

```python
# try 구문 내부에서 return 키워드 사용하는 경우

def test():
    print("test() 함수의 첫 줄입니다.")
    try:
        print("try 구문이 실행되었습니다.")
        return
        print("try 구문의 return 키워드 뒤입니다.")
    except:
        print("except 구문이 실행되었습니다.")
    else:
        print("else 구문이 실행되었습니다.")
    finally:
        print("finally 구문이 실행되었습니다.")
    print("test() 함수의 마지막 줄입니다.")
    
test()
```

```python
# finally 키워드 활용

def write_text_file(filename, text):
    try:
        file=open(filename,"w")
        return       # 파일이 열린 상태로 함수를 나가게 됨
        file.write(text)
    except Exception as e:
        print(e)
    finally:
        file.close()  # 그러나 finally 구문에 close() 함수를 입력하여 파일을 닫히게 함!
        
write_text_file("test.txt","안녕하세요!")

# try구문 중간에 return 키워드 등으로 함수를 빠져나갈 때마다 close()하도록 코드를 작성한다면 복잡해질 것
# finally는 무조건 실행이기 때문에 return이 중간에 있는 함수일 경우 close()를 finally에 써주면 코드가 깔끔
```

```python
# 반복문과 함께 사용하는 경우

print("프로그램이 시작되었습니다.")

while True:
    try:
        print("try 구문이 실행되었습니다.")
        break    # break 때문에 중간에 try구문을 빠져나감
        print("try 구문의 break 키워드 뒤입니다.") # X
    except:
        print("except 구문이 실행되었습니다.") # X
    finally:
        print("finally 구문이 실행되었습니다.")    # finally 구문 실행됨
    print("while 반복문의 마지막 줄입니다.") # X
print("프로그램이 종료되었습니다.")
```
