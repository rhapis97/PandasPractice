```python
tt = sns.load_dataset('titanic')
```

```python
tt.head()
```

```python
tt.head(10)
```

```python
tt.info()
```

```python
# 결측치가 많이 존재하는 데이터
# age, embarked, deck, embark_town
```

```python
tt.describe()
```

```python
tt['age'].mean()
```

```python
tt['age'].median()
```

```python
# 결측치 처리하는 방법: 1)삭제 2)평균이나 중위수로 대체 등
m = tt['age'].median()
```

```python
tt['age'] = tt['age'].fillna(m)    # 중위수로 결측치 대체
```

```python
tt.describe()
```

```python
# 나이에 대한 정보 개수가 채워짐
```

```python
tt.head(10)    # 결측치였던 age 부분이 중위수인 28로 대체된 모습 확인 가능
```
